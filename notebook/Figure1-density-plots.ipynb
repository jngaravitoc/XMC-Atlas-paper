```python
import numpy as np
from nba.ios import load_halo, load_snapshot
from nba.visuals import Visuals
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
plt.style.use('/mnt/home/nico/matplotlib.mplstyle')
```

```python
snapshot = "/mnt/home/nico/ceph/MWLMC_sims/MWLMC/MWLMC5_b0/MWLMC5_100M_beta0_vir_OM5_G4"

nhost=100000000
nsat=15000000
```

```python
def get_sims_hist(k):
    pos = load_snapshot(snapname=snapshot+"_{:03d}".format(k), snapformat=3, quantity='pos', ptype='dm')
    mass = load_snapshot(snapname=snapshot+"_{:03d}".format(k), snapformat=3, quantity='mass', ptype='dm')
    rsims = np.sqrt(np.sum(pos**2, axis=1))
    rcut = np.where((rsims<430))[0]


    nbins=200
    hbyz = np.histogram2d(pos[rcut,1], pos[rcut,2], bins=nbins)
    extent1 = [np.min(hbyz[1]), np.max(hbyz[1]), np.min(hbyz[2]), np.max(hbyz[2])]

    hbxz = np.histogram2d(pos[rcut,0], pos[rcut,2], bins=nbins)
    extent2 = [np.min(hbxz[1]), np.max(hbxz[1]), np.min(hbxz[2]), np.max(hbxz[2])]
    hbxy = np.histogram2d(pos[rcut,0], pos[rcut,2], bins=nbins)
    extent3 = [np.min(hbxy[1]), np.max(hbxy[1]), np.min(hbxy[2]), np.max(hbxy[2])]

    return [hbyz[0], hbxz[0], hbxy[0]], [extent1, extent2, extent3]
```

```python
h_0, ext_0 = get_sims_hist(0)
h_40, ext_40 = get_sims_hist(40)
h_70, ext_70 = get_sims_hist(70)
```

```python
fig, ax = plt.subplots(3, 3, figsize=(9, 9), sharey=True, sharex=True)

im1= ax[0][0].imshow(np.log10(np.abs(h_0[0].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_0[0], cmap='inferno', aspect='equal')
ax[0][0].set_xlim(-300, 300)
ax[0][0].set_ylim(-300, 300)

im2= ax[0][1].imshow(np.log10(np.abs(h_0[1].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_0[1], cmap='inferno', aspect='equal')
ax[0][1].set_xlim(-300, 300)
ax[0][1].set_ylim(-300, 300)

im3= ax[0][2].imshow(np.log10(np.abs(h_0[2].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_0[2], cmap='inferno', aspect='equal')
ax[0][2].set_xlim(-300, 300)
ax[0][2].set_ylim(-300, 300)


im1= ax[1][0].imshow(np.log10(np.abs(h_40[0].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_40[0], cmap='inferno', aspect='equal')
ax[1][0].set_xlim(-300, 300)
ax[1][0].set_ylim(-300, 300)

im2= ax[1][1].imshow(np.log10(np.abs(h_40[1].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_40[1], cmap='inferno', aspect='equal')
ax[1][1].set_xlim(-300, 300)
ax[1][1].set_ylim(-300, 300)

im3= ax[1][2].imshow(np.log10(np.abs(h_40[2].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_40[2], cmap='inferno', aspect='equal')
ax[1][2].set_xlim(-300, 300)
ax[1][2].set_ylim(-300, 300)

im1= ax[2][0].imshow(np.log10(np.abs(h_70[0].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_70[0], cmap='inferno', aspect='equal')
ax[2][0].set_xlim(-300, 300)
ax[2][0].set_ylim(-300, 300)

im2= ax[2][1].imshow(np.log10(np.abs(h_70[1].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_70[1], cmap='inferno', aspect='equal')
ax[2][1].set_xlim(-300, 300)
ax[2][1].set_ylim(-300, 300)

im3= ax[2][2].imshow(np.log10(np.abs(h_70[2].T)), vmin=2, vmax=5.5, origin='lower', extent=ext_70[2], cmap='inferno', aspect='equal')
ax[2][2].set_xlim(-300, 300)
ax[2][2].set_ylim(-300, 300)

fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0, hspace=0)

ax[0][0].set_title(r'$y-z$')
ax[0][1].set_title(r'$x-z$')
ax[0][2].set_title(r'$x-y$')

ax[1][0].set_ylabel(r'$\rm{kpc}$')
ax[2][1].set_xlabel(r'$\rm{kpc}$')

cbar = fig.colorbar(im1, ax=ax[0:3], orientation='horizontal',
                   cax = ax[2][0].inset_axes([0.5, -0.4, 2, 0.05]))
cbar.set_label(r'$\rm{Log_{10}}\rho_{DM}$')
```
