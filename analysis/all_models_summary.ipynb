```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("/home/ngc/Work/research/MW_viz/matplotlib.mplstyle")
```

```python
MW_GC = [1.2e12, 1.2e12, 1.2e12, 1.2e12]
LMC_GC = [8e10, 1e11, 1.8e11, 2.5e11]

MW_vasiliev = [1.0e12, 1.0e12, 1.1e12, 1.1e12]
LMC_vasiliev = [2e11, 3e11, 2e11, 3e11]

MW_lee = [7.92e11]
LMC_lee = [1.25e11]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.scatter(np.log10(MW_vasiliev), np.log10(LMC_vasiliev), label='Vasiliev+23', marker='o', color='tomato', edgecolors='k', s=60)
ax.scatter(np.log10(MW_lee), np.log10(LMC_lee), label='Lilleengen+22', marker='*', edgecolors='k', s=120, color='purple')
ax.scatter(np.log10(MW_GC), np.log10(LMC_GC), label='Garavito-Camargo+21', color='cadetblue', s=70, marker='p', edgecolors='k')
ax.set_ylabel(r'$\rm{Log_{10}\ \rm{M}_{LMC}}$')
ax.set_xlabel(r'$\rm{Log_{10}\ \rm{M}_{MW}}$')
#ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlim(np.log10(7e11), np.log10(1.5e12))
ax.set_ylim(np.log10(7e10), np.log10(4e11))
ax.set_xticks(np.arange(11.8, 12.3, 0.1))
ax.set_yticks(np.arange(10.8, 11.7, 0.2))

ax.legend(loc='lower left', fontsize=13)
plt.savefig('MWLMC_models.png', bbox_inches='tight')
plt.savefig('MWLMC_models.pdf', bbox_inches='tight')

plt.show()
```
